```python
from qutip import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
# |0>
q0 = basis(2, 0)
print(f"q_0 = {q0}")
```

```python
# (|0> + |1>) / sqrt(2)
q_diag = (basis(2, 0) + basis(2, 1)).unit()
print(f"q_diag = {q_diag}")
```

```python
b = Bloch()
b.make_sphere()
```

```python
# plotting state:
#1) just |0>
b.clear()
b.add_states(q0)

b.render()
b.show()
```

```python
# plotting state:
#2)  sigmaX |0> 
b.clear()
b.add_states(q0)

new = sigmax() * q0
print(f"new = {new}")
b.add_states(new)
b.render()
b.show()
```

```python
# plotting state:
#3)  sigmaY |0> 
b.clear()
b.add_states(q0)

new = sigmay() * q0
print(f"new = {new}")
b.add_states(new)
b.render()
b.show()
```

```python
# plotting state:
#4)  sigmaZ |0> 
b.clear()
b.add_states(q0)

new = sigmaz() * q0
print(f"new = {new}")
b.add_states(new)
b.render()
b.show()
```

```python
# plotting state:
#1) (|0> + |1>) / sqrt(2)
b.clear()
b.add_states(q_diag)

b.render()
b.show()
```

```python
# plotting state:
#2)  sigmaX * (|0> + |1>) / sqrt(2) 
b.clear()
b.add_states(q_diag)

new = sigmax() * q_diag
print(f"new = {new}")
b.add_states(new)
b.render()
b.show()
```

```python
# plotting state:
#2)  sigmaY * (|0> + |1>) / sqrt(2) 
b.clear()
b.add_states(q_diag)

new = sigmay() * q_diag
print(f"new = {new}")
b.add_states(new)
b.render()
b.show()
```

```python
# plotting state:
#3)  sigmaY * (|0> + |1>) / sqrt(2) 
b.clear()
b.add_states(q_diag)

new = sigmaz() * q_diag
print(f"new = {new}")
b.add_states(new)
b.render()
b.show()
```
